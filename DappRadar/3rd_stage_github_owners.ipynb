```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("./2nd_stage_scraping.csv")
print(df["Github"])
print(df.columns)
```

```python
df3 = pd.DataFrame(columns=["Name" , "Link" , "Github", "Owner", "Category", "Blockchain"])

for ind,row in df.iterrows():

  link = row["Github"]
  try:
    resp = requests.get(link)

  except:
    continue 
  
  if(resp.status_code!=200):
    continue

  ctr = 0
  pos3 = -1   #3rd occurence of /
  pos4 = -1   #3rd occurence of /

  for i in range(len(link)):
    if link[i]=='/':
      ctr+=1
      if ctr==3:
        pos3 = i 
      if ctr==4:
        pos4= i  
        break  

  if (pos3 == -1 and pos4 == -1) or pos3 == -1: 
    continue  
  
  if pos4 == -1:
    owner = link[pos3+1:]
  else:
    owner = link[pos3+1:pos4]
  
  if "?" in owner:
    mark_pos = owner.find("?")
    owner = owner[:mark_pos]

  print(owner)

  df2 = {"Name": row["Name"], "Link": row["Link"], "Github": link, "Owner": owner, "Category":row["Category"], "Blockchain":row["Blockchain"]}
  df3 = df3.append(df2,ignore_index=True)
  
df3.to_csv("./3rd_stage_git_owners.csv")
```
