```python
name_seen = []
for page in range(101,106):
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(f'https://dappradar.com/rankings/{page}')

    time.sleep(5)

    # Access requests via the `requests` attribute
    for request in driver.requests:
        if request.response:
            if "params" in request.url:
                r = requests.get(request.url).json()
                # print (json.dumps(r, indent=4, sort_keys=True))
                dapps = r["dapps"]
                for i in dapps:
                    name = i["name"]
                    # print(name,end=" ")
                    category = i["category"]
                    # print(category,end=" ")
                    l = len(i["protocols"])
                    # print(l,end=" ")
                    if(l == 1):
                        chain = i["protocols"][0]
                    else:
                        chain = "multichain"

                    lower_case_name = ""
                    for c in name:
                        if c.isupper():
                            lower_case_name+=c.lower()
                        elif c==" ":
                            lower_case_name+="-"
                        else:
                            lower_case_name+=c 


                    # print(chain,end=" ")  
                    link = "https://dappradar.com/"+chain+"/"+category+"/"+lower_case_name
                    # print(link)
                    df2 = {'Name':name, 'Link':link, 'Category': category, 'Blockchain':chain}
                    if name not in name_seen:
                        df = df.append(df2,ignore_index=True)
                        print(link)
                        name_seen.append(name)
                driver.close()
                break  
    
    df.to_csv("./1st_stage_scraping.csv")
```
