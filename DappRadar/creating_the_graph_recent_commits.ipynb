```python
#open the recent project csv
df = pd.read_csv("./contributors_recent_projects/recent_projects.csv")
print(df.head())
```

```python
for ind,row in df.iterrows():
  if(os.path.exists("./contributors/"+row["Name"]+".json") == False):
    print(row["Name"])

#all of the projects exists in the contributors folders
```

```python
#all contributor list
all_contributors = []

#to mark each contributor
contributor_id = {}

#project id 
project_id = {}

#map contributor ctr's to that project
contributor_map = {}

json_ctr = 0  
contributor_ctr = 0
project_ctr = 0

for name in df["Name"].tolist():
  # if(project_ctr==10):
  #   break  

  try:
    df2 = pd.read_json("./contributors/"+name+".json")
    print(name)
  except:
    print("not found "+name+".json")
    print(os.path.exists(f"./contributors/{name}.json"))
    continue

  if len(df2["contributors"])==0:
    continue

  project_ctr+=1 
  project_id[df2["id"][0]] = project_ctr

  #initialize a set to each project contributor map
  contributor_map[project_ctr] = set()

  for contributor in df2["contributors"]:
    if contributor[0] not in all_contributors:
      contributor_ctr+=1
      all_contributors.append(contributor[0])
      contributor_id[contributor[0]] = contributor_ctr
    contributor_map[project_ctr].add(contributor_id[contributor[0]])

f = open("./louvain/graph_in_figures_recent_commits.txt",'w')

for i in range(1,project_ctr):
  for j in range(i+1,project_ctr+1):
    s1 = contributor_map[i]
    s2 = contributor_map[j]
    s_intersection = s1 & s2
    print(i," ",j," ",len(s_intersection))
    print(list(s_intersection))
    f.write(str(i)+" "+str(j)+" "+str(len(s_intersection)))
    f.write("\n")

f.close()

#contributor list ---> name --> id
df2 = pd.DataFrame(contributor_id.items(),columns=["contributor name", "contributor id"])
df2.to_csv("./contributors_recent_projects/louvain/contributor_id_mapping_recent_commits.csv")

#project list --> name of project --> id
df3 = pd.DataFrame(project_id.items(),columns=["project name", "project id"])
df3.to_csv("./contributors_recent_projects/louvain/project_id_mapping.csv")
```
