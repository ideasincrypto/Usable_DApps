```python
import pandas as pd  
import json 
import os  
import matplotlib.pyplot as plt  
```

```python
all_jsons = os.listdir("./contributors/")
# print(all_jsons)
```

```python
m=dict()
for j in all_jsons:
  with open(f"./contributors/{j}") as newjson:
    jsonread = json.load(newjson)
    for con in jsonread["contributors"]:
      m.setdefault(con[0],[]).append(jsonread["id"])
```

```python
seen=[]
contributors_cnt={}
for k,v in jsonread.items():
    contrilist = v["contributorlist"]
    for contri in contrilist:
        if contri["id"] in seen:
            contributors_cnt[contri["id"]]+=1
        else:
            contributors_cnt[contri["id"]]=1
            seen.append(contri["id"])

cntlist = sorted(contributors_cnt.items(), key=lambda x:x[1])
print(len(seen))
```

```python
m2=dict()  
seen = []
for k,v in m.items():
  num_of_proj = len(v) 
  if num_of_proj in seen: 
    m2[num_of_proj]+=1 
  else:
    m2[num_of_proj]=1
    seen.append(num_of_proj)

m3 = sorted(m2.items(), key=lambda x:x[0])
print(m3)
xval = []
yval = []
for k in m3: 
  xval.append(k[0])
  yval.append(k[1])

plt.xlabel("Number of unique projects each dev has committed to")
plt.ylabel("Number of developers common")
plt.plot(xval,yval)
```
