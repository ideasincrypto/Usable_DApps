```python
import pandas as pd
import requests
import json
import time
import os
```

```python
df = pd.read_csv("./3rd_stage_git_owners.csv")
df.head()
```

```python
username = "abhilashdzr"
token = "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"
gh_session = requests.Session()
gh_session.auth = (username, token)
```

```python
common_url = "https://api.github.com"

for ind,row in df.iterrows():
    
    ownerjson = {}
    set_contributors = set()

    owner = row["Owner"]

    path = f"/home/abhilash/Desktop/August_scraping/State_Dapps/contributors/{owner}.json"
    if os.path.exists(path):
        continue
    
    print(row['Owner'])

    ownerjson["id"] = owner

    url = common_url + f'/users/{owner}/repos'
    
    page = 1
    another_page = True
    repo_list = []
    try:
        while another_page:
            gh_session.params = {'page':page, 'per-page':100}
            r = gh_session.get(url)
            # print(r.text)
            json_response = json.loads(r.text)
            for each in json_response:
                repo_list.append(each["name"])
            if 'next' in r.links:
                page +=1
                time.sleep(1)
            else:
                page = 1
                another_page = False
                
    except:
        pass
    # time.sleep(1)

    url2 = common_url + f'/users/{owner}'

    try:
        ownerjson["location"] = gh_session.get(url2).json()["location"]
    except:
        pass
    
    print(repo_list)
    # print(repo_json)
    for repo in repo_list:
        try:
            contributor_url = common_url + f'/repos/{owner}/{repo}/stats/contributors'
            contributors = gh_session.get(contributor_url).json()

            for c in contributors:
                id = c["author"]["login"]

                contributor_url = common_url + f'/users/{id}'
                this_contributor = gh_session.get(contributor_url).json()
                time.sleep(.5)

                set_contributors.add((id, this_contributor["location"]))
            
        except:
            continue


    ownerjson["contributors"] = list(set_contributors)

    json_object = json.dumps(ownerjson, indent = 4)
    f = open(f"./contributors/{owner}.json", "x")
    f.write(json_object)
    f.close()

    # time.sleep(5)
    if ind%5==0:
        time.sleep(10)
```
