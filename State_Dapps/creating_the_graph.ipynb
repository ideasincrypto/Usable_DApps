```python
all_json_name_list = os.listdir("./contributors/")
print(all_json_name_list)
```

```python
#all contributor list
all_contributors = []

#to mark each contributor
contributor_id = {}

#project id 
project_id = {}

#map contributor ctr's to that project
contributor_map = {}

json_ctr = 0  
contributor_ctr = 0
project_ctr = 0

for name in all_json_name_list:
  df = pd.read_json("./contributors/"+name)

  if len(df["contributors"])==0:
    continue

  project_ctr+=1 
  print(name)
  project_id[df["id"][0]] = project_ctr

  #initialize a set to each project contributor map
  contributor_map[project_ctr] = set()

  for contributor in df["contributors"]:
    if contributor[0] not in all_contributors:
      contributor_ctr+=1
      all_contributors.append(contributor[0])
      contributor_id[contributor[0]] = contributor_ctr
    contributor_map[project_ctr].add(contributor_id[contributor[0]])

  # if(project_ctr==6):
  #     break

f = open("./louvain/graph_in_figures.txt",'w')

for i in range(1,project_ctr):
  for j in range(i+1,project_ctr+1):
    s1 = contributor_map[i]
    s2 = contributor_map[j]
    s_intersection = s1 & s2
    print(i," ",j," ",len(s_intersection))
    f.write(str(i)+" "+str(j)+" "+str(len(s_intersection)))
    f.write("\n")

f.close()

#contributor list ---> name --> id
df2 = pd.DataFrame(contributor_id.items(),columns=["contributor name", "contributor id"])
df2.to_csv("./louvain/contributor_id_mapping.csv")

#project list --> name of project --> id
df3 = pd.DataFrame(project_id.items(),columns=["project name", "project id"])
df3.to_csv("./louvain/project_id_mapping.csv")

#graph in words 
#graph in figures (txt)
```
